```python
#adding AS to compare with plots in this data
as_df = pd.read_excel('pnas.1616408114.sd03.xlsx')
df['Accelerated Stability'] = as_df['Slope for Accelerated Stability']
df['Name'] = as_df['Name']
df.head()
```

```python
df = df[df['Name'] != 'rilotumumab']
```

```python
#scale the data - Min-Max method
from sklearn.preprocessing import MinMaxScaler
import seaborn as sn

target_columns = ['Zeta_potential', 'pI_PROPKA_based', 'pI_model_pKa_based']
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df[target_columns])
scaled_df = pd.DataFrame(scaled_values, columns=target_columns)
df[target_columns] = scaled_df
scaled_df.head()
```

```python
target_columns = ['Zeta_potential', 'pI_PROPKA_based', 'pI_model_pKa_based']
#plot scatter plots with each of the target columns with Accelerated Stability and print r squared values

for column in target_columns:
    sn.lmplot(x=column, y='Accelerated Stability', data=df)

```
